```python
print("Data Frame analysis for 'df_cash_request'")
print(f"Total amount of records in this data set: {df_cash_request.shape[0]} \n")
print(f"The data set contains {df_cash_request.shape[1]} columns:")
column_names_list = df_cash_request.columns.to_list()
print(f"{column_names_list}\n")

analysis_dict = {}
analysis_dict_row_template = {
    'unique_values': '',
    'unique_values_count': '',
    # data type analysis
    'data_type': '',
    'is_numerical': '',
    'is_categorical': '',
    'is_nominal': '',
    'is_ordinal': '',
    'is_continuous': '',
    'is_discrete': '',
    # discriptive stats
    'mean': '',
    'median': '',
    'mode': '',
    'standard deviation': '',
    'dispersion range': '',
    'frequency counts': '',
}

for column_name in df_cash_request:
    # DEBUG
    # print(f"Analysing data frame column: {column_name}")
    column = df_cash_request[column_name]

    # new empty entry in the dictionary with a fresh copy
    analysis_dict_row = analysis_dict_row_template.copy()

    analysis_dict_row['unique_values_count'] = len(column.unique())

    if analysis_dict_row['unique_values_count'] < 10:
        analysis_dict_row['unique_values'] = column.unique()
        print(f"Unique values in {column_name} are:\n {column.unique()}")

     # classify numbers from other
    if df_cash_request[column_name].dtype == 'int64':
        analysis_dict_row['data_type'] = 'integer'
        analysis_dict_row['is_numerical'] = True
    else:
        # point of improvement. dtype returns Object
        analysis_dict_row['data_type'] = 'not integer'
        analysis_dict_row['is_numerical'] = False

    analysis_dict_row['is_categorical'] = not analysis_dict_row['is_numerical']
    
    # only gender column contains nominal values
    analysis_dict_row['is_nominal'] = column_name == 'gender' or column_name == 'race/ethnicity'
    # all other text fields are ordinal
    analysis_dict_row['is_ordinal'] = analysis_dict_row['is_categorical'] and not analysis_dict_row['is_nominal'] 

    # all columns with scores contain continuous values
    analysis_dict_row['is_continuous'] = 'score' in column_name

    # set the other nominal as discrete if any
    analysis_dict_row['is_discrete'] = analysis_dict_row['is_nominal'] and not analysis_dict_row['is_continuous']
    

    if analysis_dict_row['is_numerical']:
        analysis_dict_row['mean'] = column.mean()
        analysis_dict_row['median'] = column.median()
        analysis_dict_row['mode'] = column.mode().tolist()
        analysis_dict_row['standard deviation'] = column.std()
        analysis_dict_row['dispersion range'] = column.max() - column.min()
    elif analysis_dict_row['is_categorical']:
        analysis_dict_row['frequency counts'] = column.value_counts().to_dict()
    else:
        raise TypeError(f"Failed to classify data of type {analysis_dict_row['data_type']}.")
    
    # store the result record in the report dictionary
    analysis_dict[column_name] = analysis_dict_row.copy()

separator = "----------------------------\n"
print(separator)

df_cash_request_report = pd.DataFrame(analysis_dict).transpose()
df_cash_request_report.head(20)
df_cash_request.to_pickle('df_cash_request.pkl')
df_cash_request_report.to_pickle('df_cash_request_report.pkl')

print("Data frame column types analysis report.")

print(separator)
print("The numerical variables are presented in columns:")
numerical_columns = df_cash_request_report[df_cash_request_report['is_numerical'] == True].index.tolist()
print(numerical_columns)
print("The categorical variables are presented in columns:")
categorical_columns = df_cash_request_report[df_cash_request_report['is_categorical'] == True].index.tolist()
print(categorical_columns)
print(f"Columns with nominal values:")
nominal_columns = df_cash_request_report[df_cash_request_report['is_nominal'] == True].index.tolist()
print(nominal_columns)
print(f"Columns with ordinal values:")
ordinal_columns = df_cash_request_report[df_cash_request_report['is_ordinal'] == True].index.tolist()
print(ordinal_columns)

print(separator)
print("Descriptive statistics for numerical data:")
column_names_list = ['mean', 'median', 'mode', 'standard deviation', 'dispersion range']
stats_for_numerical_columns = df_cash_request_report[df_cash_request_report['is_numerical'] == True][column_names_list]
print(stats_for_numerical_columns)
print(separator)

print("Descriptive statistics for categorical data:")
categorical_columns = df_cash_request_report[df_cash_request_report['is_categorical'] == True]['frequency counts']

for column, freq_dict in categorical_columns.items():
    print(f"\nFrequency counts for {column}:")
    freq_df = pd.DataFrame(list(freq_dict.items()), columns=['Category', 'Count'])
    print(freq_df)
```
